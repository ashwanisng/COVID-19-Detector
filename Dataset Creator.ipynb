```python
import pandas as pd
import os
import shutil
```

```python
FILE_PATH = "covid-chestxray-dataset-master/metadata.csv"
IMAGE_PATH = "covid-chestxray-dataset-master/images"
```

```python
df = pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head()
```

```python
TARGET_DIR = "Dataset/covid"
```

```python
if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("covid folder created")
```

```python
count = 0

for(i, row) in df.iterrows():
    if row['finding' ] == "COVID-19" and row['view'] == "PA":
        filename = row["filename"]
        img_path = os.path.join(IMAGE_PATH, filename)
        img_copy_path = os.path.join(TARGET_DIR, filename)
        shutil.copy2(img_path, img_copy_path)
        print("Moving Image", count)
        count += 1
print(count)
```
